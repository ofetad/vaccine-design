```python
import pandas as pd

# Load the data with a comma as the delimiter
data = pd.read_csv('filtered_MHC-I_binding_predictions.csv', delimiter=',')

# Extract the peptide column
peptides = data['peptide']

# Add sequence headers and format the output
output = []
for i, peptide in enumerate(peptides, start=1):
    header = f">Sequence_{i}"
    sequence_with_header = f"{header}\n{peptide}"
    output.append(sequence_with_header)

    # Print each sequence with header
    print(sequence_with_header)
    
# Optionally, write the output to a file
with open('peptide_sequences.txt', 'w') as file:
    for line in output:
        file.write(line + "\n")

print("Peptide sequences have been written to 'peptide_sequences.txt'.")
```
